```python
import matplotlib.pyplot as plt
from kmodes.kprototypes import KPrototypes
from sklearn.metrics import silhouette_score

# Assuming X_train is your training data
# Define the indices of the categorical columns (from 12 to 138 inclusive)
categorical_indices = [1,3,11,12,13,16]
# Initialize lists for WCSS and silhouette scores
wcss = []  # within cluster sum of squares
silhouette_scores = []

# Loop over different values of k
for k in range(2, 11):
    kproto = KPrototypes(n_clusters=k, init='Cao', n_init=1, verbose=0, random_state=28)
    clusters = kproto.fit_predict(X_train, categorical=categorical_indices)  # Specify all categorical columns

    # Calculate within-cluster sum of squares (WCSS)
    wcss.append(kproto.cost_)

    # Calculate silhouette score
    silhouette_score_val = silhouette_score(X_train, clusters)
    silhouette_scores.append(silhouette_score_val)

    print(f"Completed for k={k}")

# Plotting the elbow curve for WCSS
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), wcss, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal k (WCSS)')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')

# Plotting silhouette scores
plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-')
plt.title('Silhouette Score for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')

# Show plots
plt.tight_layout()
plt.show()
```

## Kmeans

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle

# Assuming X_train is your dataset
categorical_indices = [1, 3, 11, 12, 13, 16]

# Encode categorical columns
label_encoders = {}
for col in categorical_indices:
    le = LabelEncoder()
    X_train.iloc[:, col] = le.fit_transform(X_train.iloc[:, col])
    label_encoders[col] = le

# Convert to numpy array for faster processing
X_train = X_train.values
n_samples, n_features = X_train.shape
num_features = [i for i in range(n_features) if i not in categorical_indices]

# Custom distance function for mixed data
def custom_distance(X, Y):
    # Separate numerical and categorical features
    X_num, X_cat = X[:, num_features], X[:, categorical_indices]
    Y_num, Y_cat = Y[:, num_features], Y[:, categorical_indices]

    # Euclidean distance for numerical features
    dist_num = np.sqrt(((X_num - Y_num) ** 2).sum(axis=1))

    # Hamming distance for categorical features
    dist_cat = (X_cat != Y_cat).sum(axis=1)

    # Combine distances
    return dist_num + dist_cat

# KMeans algorithm with custom distance
def custom_kmeans(X, n_clusters, max_iter=100, tol=1e-4):
    # Initialize centroids by shuffling and picking first n_clusters
    centroids = shuffle(X)[:n_clusters]
    labels = np.zeros(n_samples)

    for iteration in range(max_iter):
        # Step 1: Assign clusters based on custom distance
        for i in range(n_samples):
            distances = custom_distance(X[i].reshape(1, -1), centroids)
            labels[i] = np.argmin(distances)

        # Step 2: Recompute centroids
        new_centroids = np.zeros_like(centroids)
        for k in range(n_clusters):
            cluster_points = X[labels == k]
            if len(cluster_points) == 0:  # Skip empty clusters
                new_centroids[k] = centroids[k]
                continue

            # Numerical mean for numerical features
            new_centroids[k, num_features] = cluster_points[:, num_features].mean(axis=0)

            # Mode for categorical features
            for idx in categorical_indices:
                new_centroids[k, idx] = np.argmax(np.bincount(cluster_points[:, idx].astype(int)))

        # Check convergence
        if np.all(np.abs(new_centroids - centroids) < tol):
            break
        centroids = new_centroids

    return labels, centroids

# Run custom KMeans for different values of k
wcss = []
silhouette_scores = []

for k in range(2, 11):
    labels, centroids = custom_kmeans(X_train, n_clusters=k)

    # Calculate WCSS (inertia-like measure)
    wcss_k = np.sum([np.min(custom_distance(X_train[i].reshape(1, -1), centroids)) for i in range(n_samples)])
    wcss.append(wcss_k)

    # Calculate silhouette score for combined data
    silhouette_score_val = silhouette_score(X_train, labels, metric='euclidean')
    silhouette_scores.append(silhouette_score_val)
    print(f"Completed for k={k}")

# Plot evaluation metrics
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), wcss, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal k (WCSS)')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')

plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-')
plt.title('Silhouette Score for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Reuse custom_distance function from above
def custom_silhouette_score(X, labels, centroids):
    silhouette_vals = []
    n_clusters = len(np.unique(labels))

    for i, x in enumerate(X):
        own_cluster = labels[i]
        own_dist = np.mean([custom_distance(x.reshape(1, -1), X[labels == own_cluster])])

        other_distances = []
        for k in range(n_clusters):
            if k != own_cluster:
                other_cluster_dist = np.mean([custom_distance(x.reshape(1, -1), X[labels == k])])
                other_distances.append(other_cluster_dist)

        closest_cluster_dist = min(other_distances) if other_distances else 0
        silhouette_val = (closest_cluster_dist - own_dist) / max(own_dist, closest_cluster_dist)
        silhouette_vals.append(silhouette_val)

    return np.array(silhouette_vals)

# Function to plot silhouette for custom KMeans
def plot_silhouette_custom(X, labels, centroids):
    silhouette_vals = custom_silhouette_score(X, labels, centroids)
    n_clusters = len(np.unique(labels))

    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.set_xlim([-0.1, 1])
    ax.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    y_lower = 10
    for i in range(n_clusters):
        cluster_silhouette_vals = silhouette_vals[labels == i]
        cluster_silhouette_vals.sort()

        size_cluster_i = cluster_silhouette_vals.shape[0]
        y_upper = y_lower + size_cluster_i

        color = plt.cm.tab10(float(i) / n_clusters)
        ax.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_vals,
                         facecolor=color, edgecolor=color, alpha=0.7)
        ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        y_lower = y_upper + 10

    silhouette_avg = silhouette_score(X[:, num_features], labels, metric='euclidean')
    ax.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax.set_xlabel("Silhouette Coefficient Values")
    ax.set_ylabel("Cluster Label")

    ax.text(silhouette_avg + 0.005, 0.5 * len(X), f'Average Silhouette Score: {silhouette_avg:.2f}', color="red")
    ax.set_title("Silhouette Plot for Clusters")
    ax.grid(True)

    plt.show()

# Call the function after clustering
labels, centroids = custom_kmeans(X_train, n_clusters=5)  # Example for 3 clusters
plot_silhouette_custom(X_train, labels, centroids)
```
